# Cleaning the data

```python
data_null=data.isnull().sum()
```

```python
data_null
```

```python
#Dropping the Redundent data
drop_deaths_data=data.drop(['Deaths','Death Ratio (%)'],axis=1)
```

```python
#Showing modified data
modified_data=drop_deaths_data.head()
```

```python
modified_data
```

# Relationship analysis

```python
corelation=drop_deaths_data.corr()
```

```python
corelation
```

```python
sns_heatmap=sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns
                       ,annot=True)
```

# Showing some plots

```python
#Pair plot
pair_plot_data=sns.pairplot(drop_deaths_data)
```

```python
#Relationship plot
replot_data=sns.relplot(x='Total Cases',y='Active',hue='State/UTs',data=drop_deaths_data)
```

```python
#Histograms
```

```python
#Hist plot
histplot_data=sns.histplot(drop_deaths_data['Total Cases'],bins=5)
```

```python
#Categoriccal plot
cat_plot_data=sns.catplot(x='Total Cases',kind='box',data=drop_deaths_data)
```
